```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numerical_df.corr(method='spearman')
display(corr_matrix)
```

```python
# @title Time_on_Social_Media vs Daily_Usage_Hours

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Time_on_Social_Media']
  ys = series['Daily_Usage_Hours']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = corr_matrix.sort_values('Time_on_Social_Media', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Time_on_Social_Media')
_ = plt.ylabel('Daily_Usage_Hours')
```

```python
# @title Daily_Usage_Hours vs Sleep_Hours

from matplotlib import pyplot as plt
corr_matrix.plot(kind='scatter', x='Daily_Usage_Hours', y='Sleep_Hours', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Age vs Daily_Usage_Hours

from matplotlib import pyplot as plt
corr_matrix.plot(kind='scatter', x='Age', y='Daily_Usage_Hours', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Age

from matplotlib import pyplot as plt
corr_matrix['Age'].plot(kind='hist', bins=20, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Time_on_Social_Media vs Sleep_Hours

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Time_on_Social_Media']
  ys = series['Sleep_Hours']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = corr_matrix.sort_values('Time_on_Social_Media', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Time_on_Social_Media')
_ = plt.ylabel('Sleep_Hours')
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

**Focus on Correlation with Target (Addiction_Level)**

```python
target_corr = corr_matrix['Addiction_Level'].drop('Addiction_Level')  # drop self-corr
target_corr = target_corr.sort_values(key=abs, ascending=False)

print("Top correlations with target:")
print(target_corr)
```

```python
threshold = 0.25
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)
high_corr = [(f1, f2, corr_pairs[(f1, f2)]) for f1, f2 in corr_pairs.index
             if f1 != f2 and abs(corr_pairs[(f1, f2)]) > threshold and f1 < f2]

for f1, f2, corr in high_corr:
    print(f"{f1} and {f2}: {corr:.2f}")
```
